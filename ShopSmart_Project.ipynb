# Feature Engineering

Time analysis is needed for this project. We need to create columns that can give more datetime information

```python

def time_analysis(x, date1, date2):
    '''
    Creates three new columns: Months, Day, and Season from the 'Date' column of the input DataFrame.

    Parameters:
    x: DataFrame containing the 'Date' column

    Returns:
    DataFrame: DataFrame with three new columns added
    '''

    # Convert 'Date' column to datetime format
    Date1 = pd.to_datetime(x[date1], format='%Y-%m-%d %H:%M:%S.%f')
    Date2 = pd.to_datetime(x[date2], format='%Y-%m-%d %H:%M:%S.%f')

    # Extracting the date
    checkin_date_ = Date1.dt.date
    checkout_date_ = Date2.dt.date

    # Extract day of the week
    checkin_day_of_week = Date1.dt.strftime('%A')
    checkout_day_of_week = Date2.dt.strftime('%A')

    # Extract time
    time_of_report = Date1.dt.strftime("%X")

    # Extract hours
    checkin_hours =  Date1.dt.hour
    checkout_hours =  Date2.dt.hour

    # Create a new DataFrame with the three new columns
    new_columns = pd.DataFrame({'checkin_date': checkin_date_, 'checkout_date':checkout_date_, 'checkin_day': checkin_day_of_week,
                                'checkout_day': checkout_day_of_week, "checkin_hour": checkin_hours, 'checkout_hour': checkout_hours})

    # Concatenate the new DataFrame with the original DataFrame x
    x = pd.concat([x, new_columns], axis=1)

    return x
```

## Visit Analysis

### Code

**Grouping and quantifing customers' visits daily and weekly**

```python
# Converting the event_timestamp to Datetime
df['event_timestamp']= pd.to_datetime(df['event_timestamp'], format='%Y-%m-%d %H:%M:%S.%f')

# Group by customer ID and date, then count the number of visits
daily_visits = df.groupby(['checkin_date', 'checkin_day', ]).size().reset_index(name='daily_visits')
daily_visits_user = df.groupby(['customer_id', 'checkin_date', 'checkin_day', ]).size().reset_index(name='daily_visits')

# Group by customer ID and week, then count the number of visits
weekly_visits = df.groupby([df['event_timestamp'].dt.to_period("W")]).size().reset_index(name='weekly_visits')
weekly_visits_user = df.groupby(['customer_id', df['event_timestamp'].dt.to_period("W")]).size().reset_index(name='weekly_visits')
```

**After grouping, let us determine the frequency**  
**We will begin with daily traffic**

```python
# Calculating average visit
avg_daily_visit = round(daily_visits['daily_visits'].mean())
avg_daily_visit_user = round(daily_visits_user['daily_visits'].mean())
min_daily =  daily_visits[daily_visits.daily_visits == daily_visits.daily_visits.min()]
max_daily =  daily_visits[daily_visits.daily_visits == daily_visits.daily_visits.max()]

printmd(f"On average, ShopSmart Inc. receives **{avg_daily_visit}** daily visits across all customers.")
printmd(f"ShopSmart Inc. recorded its least daily traffic (**{min_daily.daily_visits.values[0]}**) visits on **{min_daily.checkin_day.values[0]}, {min_daily.checkin_date.values[0]}**.")
printmd(f"Also, it recorded its highest traffic **({max_daily.daily_visits.values[0]})** on **{max_daily.checkin_day.values[0]}, {max_daily.checkin_date.values[0]}**, which made up **91.6%** of the total traffic to the e-store.")
printmd("View the chart below to see its comparison.")
print('')

# Ploting a barchart to show the distribution for those visits

plt.figure(figsize= (13,5))
plt.title('ShopSmart Inc Daily Traffic')

# Set custom colours
colours = ['#1f77b4' if x == max_daily['daily_visits'].values[0] else 'grey' for x in daily_visits.daily_visits]

sns.barplot(data= daily_visits, x='checkin_date', y= 'daily_visits', palette= colours)
plt.yscale('log')  # Set y-axis scale to logarithmic

# Add annotations
for i, visit in enumerate(daily_visits['daily_visits']):
    percent = round(visit/df.shape[0] * 100, 1)
    plt.text(i, visit + 0.5, f"{visit} ({percent}%)", ha='center', fontsize= 'x-small')

plt.xlabel('Daily Traffic')
plt.xticks(rotation=45)
plt.ylabel('Daily Traffic')
plt.show()
```

### Code

**Let us determine the location of our customers**

```python
printmd("From the map, the majority of our customers come from **Congo**, accounting for **0.76%** of our total customers. However, when grouped, it shows **Korea** took the lead, accounting for **1%**.")
printmd("The top 5 countries from which our customers come include:")
printmd(" - Korea (1.0%)")
printmd(" - Congo (0.76%)")
printmd(" - Mauritius (0.7%)")
printmd(" - Panama (0.67%)")
printmd(" - Jersey (0.67%)")
print('')

# Create a GeoDataFrame
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Getting unique customers
unique_customer = df.groupby(['customer_id', 'location']).size().reset_index()
unique_customer.columns = ['customer_id', 'location', 'na']

# Count the number of customers in each country
customer_counts = unique_customer['location'].value_counts().reset_index()
customer_counts.columns = ['country', 'customer_count']

# Calculate the percentage of customers
customer_counts['customer_percentage'] = round((customer_counts['customer_count'] / customer_counts['customer_count'].sum()) * 100, 2).astype(str) + '%'

# Merge customer counts with world data
world = world.merge(customer_counts, how='left', left_on='name', right_on='country').fillna(0)


# Create an interactive choropleth map
fig = px.choropleth(world,
                    locations='iso_a3',
                    color='customer_count',
                    hover_name='name',
                    hover_data={'name': False, 'customer_count': True, 'iso_a3': False, 'customer_percentage': True},
                    color_continuous_scale='Blues',
                    title='Customer Distribution by Country',
                    labels={'customer_count': 'Number of Customers', 'customer_percentage': 'Percentage of Customers'}
                   )

# Saving to html
fig.write_html('Customer Demographics Map.html')

fig.show()
```

```python
# Grouping each orders
unique_orders = df.groupby(['order_id', 'customer_id', 'location', 'status', 'name', 'price', 'quantity', 'event_timestamp', 'checked_out_at', ]).size().reset_index()

# I want the first record that started the order
unique_orders = unique_orders.drop_duplicates(subset=['order_id'], keep='first').reset_index()

# Removing unwanted columns
unique_orders.drop(columns= [0,'index'], inplace= True)

#Featuring an amount column which is price * quantity
unique_orders['amount'] = unique_orders['price'] * unique_orders['quantity']

# Creating time columns
unique_orders['checked_out_at'] = pd.to_datetime(unique_orders['checked_out_at'], format='%Y-%m-%d %H:%M:%S.%f')
unique_orders['event_timestamp'] = pd.to_datetime(unique_orders['event_timestamp'], format='%Y-%m-%d %H:%M:%S.%f')

unique_orders['purchase_duration'] = unique_orders['checked_out_at'] - unique_orders['event_timestamp']

# Convert the duration to hours
unique_orders['purchase_duration_hours'] = round(unique_orders['purchase_duration'].dt.total_seconds() / 3600, 2)
```
